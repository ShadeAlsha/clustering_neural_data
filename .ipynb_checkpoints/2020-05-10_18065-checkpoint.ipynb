## Introduction   
How do populations of neurons give rise to behavior? This question is at the crux of systems neuroscience research, which aims to understand the structure and function of neural circuits. As this is a largely unanswered question in the neuroscience field, in this report, we will use the tools of linear algebra to examine sub-questions that are important for begining to understand such a complex question that is how neural circuits produce behavior: 
1. Can we understand the high dimensional data of the activity of thousands of neurons in a lower dimensional subspace?
2. Can we understand the activity of neurons in terms of different functional activity profiles?  

To explore these questions, we will use a dataset provided by the Flavell lab at MIT, where I have been UROPing for the past year. The datasets comes from recordings of neurons from the roundworm C.elegans. C.elegans are an ideal model organism for studying coordianted population activity because they only have 302 neurons in their whole body, which were mapped out several decades ago. Moreover, we are able to visualize the activity of ~100 of the neurons in the head simultaneously using an imaging technique called calcium imaging.
Briefly, C.elegans are genetically engineered to express a calcium sensor called GCaMP in the neurons. A property of neurons is that when they become active, they undergo influxes of calcium, increasing the internal calcium concentration of the neuron. GCaMP changes its flourescence in the presence of calcium. Thus, the flourescent activity of GCaMP in neurons can be used as a proxy of neural activity.  
Such a technique is a neuroscientist's dream come true because it allows us to visualize the activity of many neurons simultaneously. However, as with any high dimensional measurement comes to problem of making sense of all the data. Just looking at the flourescent timeseries isn't very informative because there are tens of neurons (Problem 1). Moreover, calcium imaging doesn't tell us anything about the genetic identity of the neurons. Therefore, it is difficult to interpret the activity of the neurons based on previously studied properties of individual neurons. Fortunately, we can reduce this biological problem into one of a high dimensional matrix that doesn't have any labels. We can then use methods from linear algebra, such as dimensionality reduction and clustering, to understand the activity of the neurons in a lower dimensional space, and to identify functional "classes" of neurons based on their activity profiles. We also demonstrate how dimensionality reduction using PCA unveils clusters in the data.

## Why Dimensionality Reduction 
After some initial preprocessing of our calcium imaging dataset, we have $f_{processed}$, an `N x T` matrix, where:     
`N`: number of extracted neurons which is ~100 neurons  
`T`: time - the data is a flourescent time-series where each datapoint is the brightness of the calcium flourescent indicator (derivative with respect to time), which is used as a proxy for neural activity.  
Such a high dimensional dataset is difficult to make sense of. As we can see from the time-series graph below, simply looking at the time-series of all the neurons is not very informative. In fact, one of the big questions in systems neuroscience is how to understand the coordinated activity of large populations of neurons. Instead, dimensionality reduction may provide a better understanding of basic motifs underlying the activity of the ~100 neurons. Here, we explore in depth how Principal Component Analysis (PCA), can provide new insights into the high dimensional data C. elegans neural data.

### PCA
Principal Component Analysis, or PCA, provides a new orthogonal basis for projecting a high dimensional dataset into a lower dimensional subspace that explains most of the variance of the data. The new basis calculated by PCA is essentionaly the first `p` number of eigenvectors (where `p` sufficiently explains most of the variance), with each eigenvector providing a linear combination of the input features - in our case the ~100 neurons, or units. In doing so, we approximate $f_{processed}$ by a sum of `p` rank 1 matrices.

To perform PCA, we compute the eigendecomposition of the covariance matrix $C_{(NxN)} = \frac{f_{processed}xf_{processed}^T}{n-1}$ of our standardized data.   
The eigendecomposition of $C$ is given by:   
$ C = Q{\Lambda}Q^{-1}$, where:  
$\Lambda$: a diagonal matrix whose entries are the eigenvalues of C     
$Q$: a square matrix whose columns are the orthonormal eigenvectors. So $C$ can actually be written as  
$ C = Q{\Lambda}Q^{T}$  
We can compute the eigendecomposition of C using the `Eig` command in Julia.  

A key property of eigenvalues that is critical to understanding PCA is that:   
$ Trace(C) = \sum_{i=1}^{N}{\lambda_i}$ where ${\lambda_i}$ is the $i^{th}$ eigenvalue of C.  
Since the trace of the covariance matrix is equal to the sum of the eigenvalues of the covariance matrix, the sum of our eigenvalues is the sum of the total variance of our data (the diagonal of the covariance matrix is all the individual varainces). By ordering our eigenvectors using their corresponding eigenvalue, in order of decreasing magnitude (or decreasing variance explained), we can find a small number of principal components which explain most of the variance of our data. By projecting our data onto the first couple of principal components, we reduce the dimensionality of our ~100 dimensional dataset into `p` dimensions.

```julia
#covariance matrix
C = f_processed*f_processed'./(size(f_processed,2)-1);
#eigendecomposition of covariance matrix, C
E = eigen(C);
#eigenvalues in decreasing order of magnitude
vals = reverse(E.values);
#eigenvectors ordered in decreasing order of variance explained using corresponding eigenvalue
vectors = reverse(E.vectors,dims=2);
```

**Plotting the variance explained by the principal components - Scree Plot**.     
<p> An important question when doing dimensionality reduction using PCA is how many components are important. One method to answer this is to consider the amount of variance explained by the principal component and to identify the point at which the variance drops off using a scree plot. Below we have plotted the cumulative variance explained by the principal components (black line) and the individual variance explained (gray bars). This graph helps us determine an appropriate value for `p`. We can see that the cumulative variance starts to plataeu at ~ `p` = 5, with a total variance explained of ~65%.   
For biological data, this is an impressive amount of variance explained. In fact, we will choose a value of `p` = 2-3 as it is easier to visualize and still explains ~50% of the total variance of the data. While the variance explained of ~50% may not seem like much, in the context of biological data this is significant as there is a lot of variability in biological data because it deals with living organisms. A measurement of the same organism in the same conditions may very drastically across trials due to the inherent variability of biological data, particularly neural data.</p>

```julia

#eigenvalues proportion variance explained
val_list = (vals/sum(vals)).*100;
n = 94; #number of components to look at
PyPlot.bar(1:n,val_list[1:n],color="k", alpha=.5);
plot(1:n,cumsum(val_list[1:n]),"ko-");
ylim([0,100]);
ylabel("% Variance Explained");
xlabel("PC #");
```

**Phase Plots**   
Now that we've confirmed that the first 3 principal components explain a significant amount of the total variance in our multi-neural timeseries, we can transform $f_{processed}$ into a lower-dimensional, pxT matrix. We can do this by rotating $f_{processed}$ into the eigenvector basis. In the context of our neural timeseries dataset, each element of the eigenvector can be thought of as the contribution, or "weight" of a neuron to the corresponding principal component. Thus, the transformed data, is just a weighed sum of all the neurons for each point in time, with the weights given by the principal components. We can look at our transformed data in the PC space, with every point being a point in time going from green to blue as time progresses. Each point is a linear combination of all the neurons based on the weights from the principal component in 3D PC space. Strikingly, we find that the neural state trajectories in time form loops in PC-space, with the different "states" within PC-space visited multiple times, as indicated by the differently colored loops. These trajectories map out a sub-region within 3D PC space, which is known as a neural manifold. Thus, from our PCA analysis, we were able to identify lower dimensional components that expalin most of the variance of the data but which contain a looping neural trajectory.     

**Temporal Principal Components**: weighted sum of the full neuronal timeseries using the loadings from the principal components. These are essentially the data transformed into PC space

But how can we functionally interpret this looping trajectory? Previous work in c. elegans has suggested that the traversals of the first principal component may reflect the worm's behavior as it switches between forward and backward locomotion. We reasoned that such opposing behavior would require antagonistic neuronal groups to faciliate the switch between these antagonistic behaviors. Since our data is unlabeled, and we do not know the identities of the neurons (other than their spatial locations), we used unsupervised approaches to identify functional clusters of neurons.  

We took advantage of the fact that not only is PCA reducing the dimensionality of the data, but because our data is a timeseries, the information that is contained within the principal components is largely based on the functional relationships between the neurons. This allows us look for such functional relationships between neurons, or clusters of neurons. We further analyzed the weightings, or contribution, of each neuron to the first principal component, which explains the highest amount of variance. As discussed, above, PCA finds the principals exponents that best explain the variance of the data. But a further insight is that within each principal component, features (in our case neurons), that have similar contributions should explain similar aspects of the data [MATH PROOF on what properites could explain correlations in the weights of the first PC. What are the properties that lead to this?]. In the context of our dataset, which is composed of multi-neuronal timeseries, and of the intepretation of the first principal component as describing antagonistic behavior, we would expect that neurons with similar weightings exhibit similar functional properties (i.e correlated timeseries).
By ordering the neurons in increasing contribution to the first principal component, we immedietly see two "groups" of neurons: those that have positive and negative contributions.  

We would expect that the activity profiles of neurons within groups should have correlated activity profiles and anti-correlated activity profiles across groups. Indeed, if we reorder the neurons based on their contributions to the first principal component and compute the correlation matrices, that is exactly what we see. This is a very exciting result of conducting the PCA because it allows us to identify correlations within the data without the need for signal processing techniques. To emphasize this, we can see that just looking at the correlations between all the neurons is not informative as it is hard to see the relationships between the neurons. But using the space of the first principal component provides us with another space to view such correlation and allows us to identify basic structures in the data.
To verify that these relationships make sense, we look at the timeseries of the neurons from the two groups and find strong correlations. These opposing neurons have great biological importance as their alternating activity is thought to regulate opposing behaviors - such as forward and backward locomotion. However, it would be naiive to expect there to only be two neuronal functional groups. If we plot the first two principal components against each other, we find that the two groups (positive and negative contributions based on the heatmap) cluster well. We next sought to identify remaining functional clusters. Our first pass at clustering was based on setting binary thresholds in the heatmap, but given that the first component seemed to allow us to represent correlations, we wanted to see whether we could extract more fine-grained clusters using a more systematic method.     

N.B: It is important to emphasize that the similar contributions only relate to obviously correlated activity of the neurons when looking at the first principal component because that component likely represents opposing activity. However, in more general terms, the meaning of the similar contributions in each principal component  depends on what the component is representing. For example, if we use the same order from the first principal component to order the second principal component, we can see that in the second PC space, those neurons aren't correlated. This of course makes sense because PCA wouldn't be explaining the most variance of the dataset if every principal component represented similar things (in our case, correlations between the timeseries). So neurons with similar weight contributions in the second princpal component may be contributing to similar features, but that does not necessarily mean that the activity needs to be correlated. So the exciting  result that the first principal component extracted correlated. Thus, this functional clustering is a result of not only the inherent properties of PCA, but also of the data that we are analyzing.
